# Machine Learning

```python
stats.head()
```

### Predict total from other stats (ideally, it should be sum of stats with weight 1 except id with weight 0)

```python
X = stats[['id','hp','attack','defense','sp. atk','sp. def','speed']]
y = stats['total']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

## Convert type to integer

```python
num = df.drop(['id','name','total'],axis=1)
num.head()
```

```python
num['type 2'].unique()
```

```python
def tc(x):
    l = ['Poison', 'None', 'Flying', 'Ground', 'Fairy', 'Grass', 'Fighting',
       'Psychic', 'Steel', 'Ice', 'Rock', 'Water', 'Electric', 'Fire',
       'Dragon', 'Dark', 'Ghost', 'Bug', 'Normal']
    return l.index(x)+1
```

```python
tc('Normal')
```

```python
num['t1'] = num['type 1'].apply(tc)
num['t2'] = num['type 2'].apply(tc)
num.head()
```

```python
num.drop(['type 1', 'type 2'],axis=1,inplace=True)
num.head()
```

```python
num['legendary'].head()
```

```python
num['legendary'].apply(lambda x: 1 if(x) else 0).head()
```

```python
num['legendary']=num['legendary'].apply(lambda x: 1 if(x) else 0)
```

### Predict sp. atk

```python
X = num[['hp','attack','defense','sp. def','speed','generation','legendary','t1','t2']]
y = num['sp. atk']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
lm.fit(X_train,y_train)
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

### Predict legendary

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
X = num[['hp','attack','defense','sp. atk','sp. def','speed','generation','t1','t2']]
y = num['legendary']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
print(classification_report(y_test,predictions))
```
